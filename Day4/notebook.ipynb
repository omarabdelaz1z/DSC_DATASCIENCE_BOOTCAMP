```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
dataset_path = r'./Orange_Telecom_Churn_Data.csv'
df = pd.read_csv(dataset_path)
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# How many users on Voice Mail Plan?


```python
voice_mail_users = df[df.voice_mail_plan == 'yes']
voice_mail_users
```

```python
print(f"Number of users who use Voice Mail: {voice_mail_users.shape[0]}")
```

# What is the average of vmail messages for subscribed users only?

```python
average_vmail_messages = voice_mail_users.number_vmail_messages.mean()
print(f"Average number of voice mail messages -> {int(average_vmail_messages)} messages.")
```

# Total daylight minutes per state

```python
total_daylight_minutes_per_state = df.groupby('state')[['total_day_minutes']].agg('sum')
total_daylight_minutes_per_state
```

# Total daylight charge per state

```python
total_daylight_charge_per_state = df.groupby('state')[['total_day_charge']].agg('sum')
total_daylight_charge_per_state
```

# Average call duration in day, eve and night 

```python
df[['total_day_minutes', 'total_eve_minutes', 'total_night_minutes']].mean()
```

```python
average_call_duration_per_state = df.groupby('state')[['total_day_minutes', 'total_eve_minutes', 'total_night_minutes']].agg('mean')
average_call_duration_per_state
```

# Total day minutes and charge (day, eve and night)

```python
df[['total_day_charge', 'total_eve_charge', 'total_night_charge']].sum()
```

```python
total_charge_per_state = df.groupby('state')[['total_day_charge', 'total_eve_charge', 'total_night_charge']].agg('sum')
total_charge_per_state
```
